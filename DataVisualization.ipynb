```python
from string import ascii_letters, digits
import datetime
```

```python
password = ["".join(np.random.choice(list(ascii_letters + digits), 8))for x in range(3)]

creation_date = [datetime.datetime.now().date() - datetime.timedelta(int (x)) for x in np.random.randint(0,1500,3)]

username = ["alice26", "bob_smith", "eve"]
```

```python
print(password)

print(creation_date)

print(username)
```

```python
diss = pd.DataFrame({"Username" : username, "Password" : password, "Creation Date" : pd.to_datetime(creation_date)})

diss
```

```python
diss

emails = ["hampo@gmail.com", "grace@ymail.com", "kelvin@rocketmail.com"]

diss["email"] = emails

diss
```

```python
diss.drop(0)
diss.drop("email", axis=1)
```

```python
csv = [",".join(map(lambda x: str(x), row)) for row in np.vstack([diss.columns, diss])]
with open('.read_csv_example.csv', 'w') as f:
    [f.write(line + '\n') for line in csv]

#!cat .read_csv_example.csv
```
